# Tasks
The main aim of these tasks is to get familiar with Linear Regression problems. Apply LinearRegression to the following datasets. Note the error obtained and coefficients
*   **dataset1.csv** - a synthetic dataset including 5 points (1-dimensional data) with one target
*   **dataset2.csv** - a synthetic dataset including 1000 points (3-dimensional data) with one target variable. The variables are named x1, x2, x3, y. y is the variable to be regressed. Split the dataset into training set and test set (70% for training set and 30% for test set). 


# Task 2.1. dataset1.csv


```python
#code
data = pd.read_csv('dataset1.csv')
X_train = data[['x']]
Y_train = data[['y']]
print(X_train)
print(Y_train)
```

```python
#To create  Instance of Simple  Linear Regression Model
regressor = LinearRegression()

#To fit the X_train and Y_train
regressor.fit(X_train, Y_train)
print(regressor.intercept_)
print(regressor.coef_)
```

```python
#y = 0.4 + 0.8x
y_pred = regressor.predict(X_train)
y_pred
```

```python
#r2
r2 = metrics.r2_score(Y_train, y_pred)
r2 
```

# Task 2.2. dataset2.csv


```python
#code
dataset2 = pd.read_csv('dataset2.csv')
dataset2
x = dataset2[['x1', 'x2', 'x3']]
y = dataset2[['y']]
X2_train, X2_test, Y2_train, Y2_test  = train_test_split(x, y, train_size=0.7, random_state=1)
X2_test
```

```python
#To create  Instance of Simple  Linear Regression Model
regressor = LinearRegression()

#To fit the X_train and Y_train
regressor.fit(X2_train, Y2_train)
print(regressor.intercept_)
print(regressor.coef_)

#y = 4.00001817 + 2.99999344x1 + 3.99999127x2 + 4.99999257x3
Y2_pred = regressor.predict(X2_test)
Y2_pred
```

```python
print(metrics.explained_variance_score(Y2_test, Y2_pred))
print(metrics.mean_absolute_error(Y2_test, Y2_pred))
print(metrics.mean_squared_error(Y2_test, Y2_pred))
print(metrics.median_absolute_error(Y2_test, Y2_pred))
print(metrics.r2_score(Y2_test, Y2_pred))
```

# Task 2.3. Apply LinearRegression to the following dataset:
*   Red wine quality: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv . The dataset includes info about the chemical properties of different types of wine and how they relate to overall quality. 



```python
#code
dataset3 = pd.read_csv('Task3_winequality-red.csv', sep=';')
dataset3
x3 = dataset3[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
y3 = dataset3[['quality']]
X3_train, X3_test, Y3_train, Y3_test = train_test_split(x3, y3, train_size=0.7, random_state=1)
X3_test
```

```python
#To create  Instance of Simple  Linear Regression Model
regressor = LinearRegression()

#To fit the X_train and Y_train
regressor.fit(X3_train, Y3_train)
print(regressor.intercept_)
print(regressor.coef_)

#y = w0 + w1x1 + w2x2 + w3x3 + ...
Y3_pred = regressor.predict(X3_test)
Y3_pred
```

```python
print(metrics.explained_variance_score(Y3_test, Y3_pred))
print(metrics.mean_absolute_error(Y3_test, Y3_pred))
print(metrics.mean_squared_error(Y3_test, Y3_pred))
print(metrics.median_absolute_error(Y3_test, Y3_pred))
print(metrics.r2_score(Y3_test, Y3_pred))
```
